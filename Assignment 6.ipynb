**b)** Find a quadrature formula of the form

$$
   \int_0^1 \,\frac{1}{\sqrt{x}}\; f(x)\,dx  = A_1 f(x_1) + A_2 f(x_2)
$$

with an optimal degree of precision. Use this formula to find an approximation to the
integral

$$
    \int_0^1 \sqrt{x}\,\sin x \; dx
$$

***Solution***: The zeros of $\varphi_2$ are $x_{1,2}=\frac37\mp\frac{2\sqrt{30}}{35}$. 
This gives the weights 

$$
    A_1 = \int_{0}^{1} x^{-1/2} \frac{x-x_2}{x_1-x_2} \,\mathrm{d}x = 1+\frac{\sqrt{30}}{18}
$$

and 

$$
    A_2 = \int_{0}^{1} x^{-1/2} \frac{x-x_1}{x_2-x_1} \,\mathrm{d}x = 1-\frac{\sqrt{30}}{18}.
$$

This gives the Gaussian quadrature

$$
    \int_0^1 \sqrt{x}\,\sin x \; dx \approx A_1\sin x_1 + A_2\sin x_2 \approx 0.36584953.
$$
