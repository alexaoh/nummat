### Problem 1
In the lectures, you learned about normal matrices, those which have a diagonalization of the form

$$
     A = U\Lambda U^H
$$

where $\Lambda$ is a diagonal complex $n\times n$-matrix and $U$ is a unitary (complex) matrix such that $U^H U=I$
(recall that $U^H$ is the complex conjugate of $U^T$).

Show that for any such matrix, one has $\|A\|_2=\rho(A)$ (where $\rho(A)$ is the spectral radius of $A$).

### Solution
Every vector norm induces a norm on matrices $V \in \mathbb{R}^n$, $A \in \mathbb{R}^{n x n}$

\begin{equation}\label{induced_norm}
    ||A|| = \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||}{||\mathbf{x}||}.
\end{equation}

It is also known that 

\begin{equation}\label{vector2norm}
    ||\mathbf{x}||_2 = \left(\sum_{i = 1}^{n} |x_i|\right)^{1/2}.
\end{equation}

A hermitian matrix $B^HB$ can be written as 
\begin{equation}
    U^HB^HBU = \text{diag}(\mu_1, \mu_2, ..., \mu_n), 
\end{equation}

where $\mu_i$ are the positive eigenvalues of $B^HB$ and $U$ is a unitary matrix.  Here, $A = U\Lambda U^H$ is hermitian, since $A^H = A$, and $B^HB$ can be exchanged for the hermitian matrix $A$ in the expression above. This means that $\mu_i$ are the positive eigenvalues of $A$, which means that $\Lambda$ is given by 

\begin{equation}
    \Lambda = \text{diag}(\lambda_1, ..., \lambda_n), 
\end{equation}

where $\lambda_i$ are the eigenvalues of $A$.

Beginning with the first equation above for the spectral norm

\begin{align*}
    ||A||_2 &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||_2}{||\mathbf{x}||_2} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A\mathbf{x},A\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A^HA\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(U\Lambda^2 U^H\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(UU^H\mathbf{x}, \mathbf{x})}} &&\text{(Since $U$ is unitary)} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(U^H\mathbf{x}, U^H\mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} &&\text{(Substitute $\mathbf{y} = U^H\mathbf{x}$)} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{\Lambda^2( \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} \\
    &= \max\limits_{i = 1, 2,\ldots, n} \sqrt{\lambda_i^2} \\
    &= \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| \\
    &= \rho (A) &&\text{$\blacksquare$} 
\end{align*}

### Solution

**a)** The eigenvalues of $A$ are given by $(-\lambda)^n = 0 \implies \lambda = 0$. This eigenvalue has multiplicity $n$. The Gershgorin circle theorem tells us that all the eigenvalues of $A$ are within at least one of the Gershgorin discs of $A$. A Gershgorin disk $D_i$ is a closed disk with centre given by the diagonal entry in row i, $a_{ii}$, and radius $r_i$ given by the sum of all the entries in row $i$ which are not on the diagonal: $ r_i = \sum_{i \neq j}a_{ij}$. In this specific case this means that all the eigenvalues of $A$ would be contained in a disk with centre 0 and radius 1. 

Since all the centres are 0 it means that $|\lambda_i| \leq r_i ,\: i = 1, 2,\ldots, n \implies |\lambda_i| \leq 1 ,\: i = 1, 2,\ldots, n$.

**b)** After the new addition to the matrix $A$, $\hat A$ takes the form

$$
\hat A = \left[
\begin{array}{rrrr}
0 & \ldots & \ldots &  \epsilon \\
1 & 0 &   & \vdots \\
\vdots & \ddots &\ddots & \vdots \\
0 & \ldots& 1 & 0 & 
\end{array}
\right].
$$

This leads to different eigenvalues than for the matrix $A$. Computing the eigenvalues as normally gives

$$
    \text{det}(\hat A - \lambda I_n) = 0 \implies (-\lambda)^n + (-1)^{2n-1}\epsilon = 0, 
$$

where $I_n$ is the identity matrix in $\mathbb{R}^n$. This ultimately gives 

$$
    \lambda = (-1)^{n-1}\epsilon^{1/n}.
$$

By the definition of the spectral radius, this gives 

$$
    \rho(\hat A) = \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| = \epsilon^{1/n}.
$$

As stated earlier an expression for the eigenvalues of $\hat A$ is 

$$
    \lambda = (-1)^{n-1}\epsilon^{1/n}, 
$$

and an expression for the eigenvectors is

$$
    (\hat A -I_n \lambda)x = 0 \implies \ldots
$$

**c)**

$$
 K_2(\hat A) = \frac{\sigma_1(\hat A)}{\sigma_n(\hat A)}, 
$$

where $\sigma_1(\hat A)$ and $\sigma_n(\hat A)$ are the maximum and minimum singular values of $\hat A$. If $\hat A$ is positive definite, this leads to the definition

$$
    K_2(\hat A) = \frac{\lambda_{max}}{\lambda_{min}} = \rho(\hat A)\rho(\hat A).
$$

Therefore, it is very easy to find $K_2(\hat A)$ if it fulfills these requirements.
