### Problem 1
In the lectures, you learned about normal matrices, those which have a diagonalization of the form

$$
     A = U\Lambda U^H
$$

where $\Lambda$ is a diagonal complex $n\times n$-matrix and $U$ is a unitary (complex) matrix such that $U^H U=I$
(recall that $U^H$ is the complex conjugate of $U^T$).

Show that for any such matrix, one has $\|A\|_2=\rho(A)$ (where $\rho(A)$ is the spectral radius of $A$).

### Solution
Every vector norm induces a norm on matrices $V \in \mathbb{R}^n$, $A \in \mathbb{R}^{n x n}$

$$
\begin{equation}\label{induced_norm}
    ||A|| = \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||}{||\mathbf{x}||}.
\end{equation}
$$

It is also known that 

$$
\begin{equation}\label{vector2norm}
    ||\mathbf{x}||_2 = \left(\sum_{i = 1}^{n} |x_i|\right)^{1/2}.
\end{equation}
$$

A hermitian matrix $B^HB$ can be written as 

$$
\begin{equation}
    U^HB^HBU = \text{diag}(\mu_1, \mu_2, ..., \mu_n), 
\end{equation}
$$

where $\mu_i$ are the positive eigenvalues of $B^HB$ and $U$ is a unitary matrix.  Here, $A = U\Lambda U^H$ is hermitian, since $A^H = A$, and $B^HB$ can be exchanged for the hermitian matrix $A$ in the expression above. This means that $\mu_i$ are the positive eigenvalues of $A$, which means that $\Lambda$ is given by 

$$
\begin{equation}
    \Lambda = \text{diag}(\lambda_1, ..., \lambda_n), 
\end{equation}
$$

where $\lambda_i$ are the eigenvalues of $A$.

Beginning with the first equation above for the spectral norm

$$
\begin{align*}
    ||A||_2 &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||_2}{||\mathbf{x}||_2} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A\mathbf{x},A\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A^HA\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(U\Lambda^2 U^H\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(UU^H\mathbf{x}, \mathbf{x})}} &&\text{(Since $U$ is unitary)} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(U^H\mathbf{x}, U^H\mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} &&\text{(Substitute $\mathbf{y} = U^H\mathbf{x}$)} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{\Lambda^2( \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} \\
    &= \max\limits_{i = 1, 2,\ldots, n} \sqrt{\lambda_i^2} \\
    &= \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| \\
    &= \rho (A) &&\text{$\, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \,\, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \blacksquare$} 
\end{align*}
$$

### Solution
 
**a)** By calculating $\text{det}(A - \lambda I_n) = 0$ one finds that the eigenvalues of $A$ are given by $(-\lambda)^n = 0 \implies \lambda = 0$. This eigenvalue has multiplicity $n$. The Gershgorin circle theorem tells us that all the eigenvalues of $A$ are within at least one of the Gershgorin discs of $A$. A Gershgorin disk $D_i$ is a closed disk with centre given by the diagonal entry in row i, $a_{ii}$, and radius $r_i$ given by the sum of all the entries in row $i$ which are not on the diagonal: $ r_i = \sum_{i \neq j}a_{ij}$. In this specific case this means that all the eigenvalues of $A$ would be contained in a disk with centre 0 and radius 1 and/or in a disk with centre 0 and radius 0. Since at least one of the eigenvalues of the matrix is contained in the disk with centre 0 and radius 0, at least one eigenvalue must be zero and $A$ must be singular. This way we could have reasoned that $A$ must be singular based on Gershgorin's theorem instead of calculating the eigenvalues of the marix, as we did to begin with.

Since all the centres are $a_ {ii} = 0$ it means that $|\lambda_i - a_{ii}| = |\lambda_i| \leq r_i ,\: i = 1, 2,\ldots, n \implies |\lambda_i| \leq 1 ,\: i = 1, 2,\ldots, n$. This corresponds with the earlier stated eigenvalues of $A$.

**b)** After the new addition to the matrix $A$, $\hat A$ takes the form

$$
\hat A = \left[
\begin{array}{rrrr}
0 & \ldots & \ldots &  \epsilon \\
1 & 0 &   & \vdots \\
\vdots & \ddots &\ddots & \vdots \\
0 & \ldots& 1 & 0 & 
\end{array}
\right].
$$

This leads to different eigenvalues than for the matrix $A$. Computing the eigenvalues as normally gives

$$
    \text{det}(\hat A - \lambda I_n) = 0 \implies (-\lambda)^n + (-1)^{n-1}\epsilon = 0, 
$$

where $I_n$ is the identity matrix in $\mathbb{R}^n$. This ultimately gives 

$$
    \lambda^n = \epsilon. 
$$

This means that all the eigenvalues of $\hat A$ are the nth complex roots of epsilon. By the definition of complex roots one finds that $|\lambda| = \epsilon^{1/n}$.  

Moreover, by the definition of the spectral radius, this gives 

$$
    \rho(\hat A) = \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| = \epsilon^{1/n}.
$$

Now that the expression for the eigenvalues has been found, an expression for the eigenvectors is given by

$$
    (\hat A -I_n \lambda_i)\mathbf{x_i} = 0, 
$$

for every $i = 1, \ldots, n$. This results in the following system

$$
\left[
\begin{array}{rrrr}
-\lambda_i & \ldots & \ldots &  \epsilon \\
1 & -\lambda_i &   & \vdots \\
\vdots & \ddots &\ddots & \vdots \\
0 & \ldots& 1 & -\lambda_i & 
\end{array}
\right] \mathbf{x_i} = 0, 
$$

where $\mathbf{x_i} = [x_{i1}, x_{i2}, \ldots, x_{in}]^T \neq 0$ is eigenvector number $i$.

This system gives the following expressions for the coordinates in $\mathbf{x}$

$$
\begin{align*}
x_1 &= \frac{\epsilon}{\lambda_i}x_n \\
x_2 = \frac{x_1}{\lambda_i} &= \frac{\epsilon}{\lambda_i^2}x_n \\
\vdots \\
x_{n-1} = \frac{x_{n-2}}{\lambda_i} &= \frac{\epsilon}{\lambda_i^{n-1}}x_n \\
x_n = \frac{x_{n-1}}{\lambda_i} &= \frac{\epsilon}{\lambda_i^n}x_n = x_n. \\
\end{align*}
$$

It is apparent that the last coordinate $x_n$ can be chosen freely. Set $x_n = \psi \in \mathbb{C}$.
Therefore, the eigenvalue $\lambda_i$ has the corresponding eigenvector $x_i$, given by  

$$
\mathbf{x_i} = \left[
\begin{array}{rrrrr}
x_1 \\ x_2 \\ x_3 \\ \vdots \\ x_n
\end{array}
\right] = 
\left[
\begin{array}{rrrr}
\lambda_i^{-1} \\ \lambda_i^{-2} \\ \lambda_i^{-3} \\ \vdots \\ \epsilon^{-1}
\end{array}
\right]\epsilon \psi,
$$

for every $i = 1, \ldots, n$. This is the expression for the eigenvectors of $\hat A$.

**c)**
It is known that $K_2(\hat{A})=\|\hat{A}\|_2\cdot\|\hat{A}^{-1}\|_2 = \frac{\sigma_1(\hat A)}{\sigma_n(\hat A)}$, where $\sigma_1{(\hat A)}$ and $\sigma_n{(\hat A)}$ are the maximum and minimum singular values of $\hat A$. The singular values are defined as $\sigma_i{(\hat A)} = \sqrt{\lambda_i (\hat A\hat A^H)}$, for $i = 1, \ldots, p$. The product $\hat A \hat A^H$ is given by

$$
\hat A \hat A^H= \left[
\begin{array}{rrrr}
0 & \ldots & \ldots &  \epsilon \\
1 & 0 &   & \vdots \\
\vdots & \ddots &\ddots & 0 \\
0 & \ldots& 1 & 0 & 
\end{array}
\right]\left[
\begin{array}{rrrr}
0 & 1 & \ldots &  0 \\
\vdots & 0 & \ddots & \vdots \\
\vdots &  & \ddots & 1 \\
\epsilon & \ldots & 0 & 0 & 
\end{array}
\right] = \left[
\begin{array}{rrrr}
\epsilon^2 & 0 & \ldots &  0 \\
0 & 1 & \ddots & \vdots \\
\vdots &  & \ddots & 0 \\
0 & \ldots & 0 & 1 & 
\end{array}
\right].
$$

Since $\hat A \hat A^H$ is a diagonal matrix, the eigenvalues are given by the diagonal entries of the matrix. When $\epsilon < 1$, the maximum eigenvalue of $\hat A\hat A^H$ is $1$ and the minimum eigenvalue is $\epsilon^2$. Ultimately, this means that 

$$
    K_2(\hat{A}) = \frac{\sigma_1(\hat A)}{\sigma_n(\hat A)} = \frac{\sqrt{1}}{\sqrt{\epsilon^2}} = \frac{1}{\epsilon}.
$$

Define the expression 

$$
    \Xi_i = \frac{\frac{\|\delta x\|_i}{\|x\|_i}}{\frac{\|\delta H_n\|_i}{\|H_n\|_i}+\frac{\|\delta b\|_i}{\|b\|_i}},
$$

for $i = 1, 2, \infty$.



```python
import numpy.random as r
import numpy.linalg as l

# Using the same n-values as in a)
n = np.arange(2, 14, 1, dtype=int)

hmin = lambda n: 1/(2*n+1) # Smallest value in H relative to all the values of n.
pert_amount = 2000 # Amount of perturbations.

# Empty dictionary to store values per value of n.
simulations = {}

for i in range(n[-2]):
    s = n[i]
    b = np.ones(s)
    H = la.hilbert(s)
    x = l.solve(H,b)
    values = np.zeros((pert_amount, 4))
    
    c = 0.001 # Factor that scales the perturbations. 
    j = hmin(s)*c
    
    for p in range(pert_amount):
        dH = r.rand(s,s)*j 
        db = r.rand(s)*j 
        x_dx = l.solve(H+dH, b+db)
        dx = x - x_dx # Find perturbation in x.

        frob_norm = (l.norm(dx)/l.norm(x))/(((l.norm(dH))/l.norm(H))+((l.norm(db)/l.norm(b))))
        inf_norm = (l.norm(dx, np.inf)/l.norm(x, np.inf))/(((l.norm(dH, np.inf))/l.norm(H, np.inf))+((l.norm(db, np.inf)/l.norm(b, np.inf))))
        one_norm = (l.norm(dx, 1)/l.norm(x, 1))/(((l.norm(dH, 1))/l.norm(H, 1))+((l.norm(db, 1)/l.norm(b, 1))))
        
        # Isn't the comparison given by dx, since that is the difference from the solution?
        val = [l.norm(dx), frob_norm, inf_norm, one_norm]
        values[p, :] = val
    
    simulations[i] = values # One list of values for ten perturbations, per n.
    

# Used to plot some averages below.
average_diffs = np.zeros(len(n))
average_frob_norm = np.zeros(len(n))
average_inf_norm = np.zeros(len(n))
average_one_norm = np.zeros(len(n))
    
def find_average_values(log = False):
    """Print values from the simulations if 'write' == True. Fill average of fractions and diffs per n"""
    for key, arr in simulations.items():
        if log:
            print("Sim for n =", n[key], ":")

        # Used if you want to print values from each perturbation per n
        if log: 
            for a in arr:
                print("||dx|| = ", a[0], ",fraction = ", a[1])
                print("\n")

        # Sum for plotting averages for each n.
        sum_diffs = np.sum(arr, axis=0)[0]
        sum_frob_norm = np.sum(arr, axis=0)[1]
        sum_inf_norm = np.sum(arr, axis = 0)[2]
        sum_one_norm = np.sum(arr, axis=0)[3]
        
        average_diffs[key] = sum_diffs/pert_amount
        average_frob_norm[key] = sum_frob_norm/pert_amount
        average_inf_norm[key] = sum_inf_norm/pert_amount
        average_one_norm[key] = sum_one_norm/pert_amount
        
        if log:
            print("\n\n")


# The print capabilities could probably be deleted!
find_average_values()

fig, ax = plt.subplots(1,2, figsize=(15, 5))
fig.suptitle("Some Averages Over "+str(pert_amount)+" Perturbations per Value of $n$")
ax[0].plot(n, average_diffs)
ax[0].set_title("Average of $||\delta x||_2$ Over Perturbations as Function of $n$")
ax[0].set_xlabel("Dim [n]")
ax[0].set_ylabel(r"$\overline{||\delta x||_2}$")
ax[1].plot(n, average_frob_norm, label="$\Xi_2$")
ax[1].plot(n, average_inf_norm, label="$\Xi_{\infty}$")
ax[1].plot(n, average_one_norm, label="$\Xi_1$")
ax[1].set_title("$\Xi_i, \, \, \, i = 1, 2, \infty$, as Function of $n$")
ax[1].set_xlabel("Dim [n]")
ax[1].set_ylabel("$\Xi_i, \, \, \, i = 1, 2, \infty$")#+ r"$\overline{\frac{\frac{\|\delta x\|}{\|x\|}}{\frac{\|\delta H_n\|}{\|H_n\|}+\frac{\|\delta b\|}{\|b\|}}}$")
ax[1].legend(loc=1)
plt.show()

#print("Some test values for the Frobenius norm:")
#print("Average of fractions for n = 2:", average_frob_norm[0])
#print("Average of fractions for n = 13:", average_frob_norm[-1])
#print("Average of differences for n = 2:", average_diffs[0])
#print("Average of differences for n = 13:", average_diffs[-1])

# I am wondering about this comparison between the solutions of the exact system and the perturbed one
# Do you want me to compare and plot ||dx|| as I have done, or is the comparison the condition number?
# If you only are interested in the condition number I should plot both the log and the regular one, as in a)!
```

Let $A$ be any invertible $n\times n$-matrix. Suppose that $\delta A$ is the smallest possible matrix, measured in a subordinate (natural) matrix norm $\|\cdot\|$ such that $A+\delta A$ is singular. Show that

$$
    \|\delta A\|  = \|A^{-1}\|^{-1}
$$



### Solution

Since $A+\delta x$ is singular, the following must hold for some $x \neq 0$

$$
    (A+\delta A)x = 0 \iff Ax + \delta Ax = 0 \iff -\delta Ax = Ax. \iff x = -A^{-1}\delta Ax, 
$$

where the last equality holds since $A$ is invertible. 

Taking the norm of this result one gets

$$
    \|x\| = \|-A^{-1}\delta Ax\| \leq \|A^{-1}\|\delta Ax\| \leq \|A^{-1}\|\delta A\|\|x\|, 
$$

where we assume that the matrix norm has the sub-multiplicative property $\|AB\| \leq \|A\|\|B\|$ and that the matrix norm and the vector norm are compatible $\|Ax\| \leq \|A\|\|x\|$. The relation therefore is

$$
    \|A^{-1}\|^{-1} \leq \|\delta A\|.
$$

Since we know that $\delta A$ is the smallest possible matrix, measured in a subordinate matrix norm such that $A+\delta A$ is singular, the equality must hold and be precise. Finally, we end up in 

$$
    \|A^{-1}\|^{-1} = \|\delta A\|, 
$$
