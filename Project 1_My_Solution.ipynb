### Problem 1
In the lectures, you learned about normal matrices, those which have a diagonalization of the form

$$
     A = U\Lambda U^H
$$

where $\Lambda$ is a diagonal complex $n\times n$-matrix and $U$ is a unitary (complex) matrix such that $U^H U=I$
(recall that $U^H$ is the complex conjugate of $U^T$).

Show that for any such matrix, one has $\|A\|_2=\rho(A)$ (where $\rho(A)$ is the spectral radius of $A$).

### Solution
Every vector norm induces a norm on matrices $V \in \mathbb{R}^n$, $A \in \mathbb{R}^{n x n}$

\begin{equation}\label{induced_norm}
    ||A|| = \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||}{||\mathbf{x}||}.
\end{equation}

It is also known that 

\begin{equation}\label{vector2norm}
    ||\mathbf{x}||_2 = \left(\sum_{i = 1}^{n} |x_i|\right)^{1/2}.
\end{equation}

A hermitian matrix $B^HB$ can be written as 
\begin{equation}
    U^HB^HBU = \text{diag}(\mu_1, \mu_2, ..., \mu_n), 
\end{equation}

where $\mu_i$ are the positive eigenvalues of $B^HB$ and $U$ is a unitary matrix.  Here, $A = U\Lambda U^H$ is hermitian, since $A^H = A$, and $B^HB$ can be exchanged for the hermitian matrix $A$ in the expression above. This means that $\mu_i$ are the positive eigenvalues of $A^2$, which means that $\Lambda$ is given by 

\begin{equation}
    \Lambda = \text{diag}(\lambda_1^2, ..., \lambda_n^2), 
\end{equation}

where $\lambda_i$ are the eigenvalues of $A$. FORTSATT NOE SOM SKURRER HER!!!

Beginning with the first equation above for the spectral norm

\begin{align*}
    ||A||_2 &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||_2}{||\mathbf{x}||_2} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(U\Lambda U^H\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda U^H\mathbf{x},U^H\mathbf{x})}{(UU^H\mathbf{x}, \mathbf{x})}} &&\text{(Since $U$ is unitary)} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda U^H\mathbf{x},U^H\mathbf{x})}{(U^H\mathbf{x}, U^H\mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} &&\text{(Substitute $\mathbf{y} = U^H\mathbf{x}$)} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{\Lambda( \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\Lambda} \\
    &= \rho (A) &&\text{$\blacksquare$} 
\end{align*}

The last equality needs to be explained. Since the eigenvalues on the diagonal of $\Lambda$ are the eigenvalues of $A^2$, the largest value among the square root of the diagonals of $\Lambda$ must be the largest eigenvalue of $A$. This is, per definition, the spectral radius $\rho (A)$ of $A$, which is what was to be shown.
