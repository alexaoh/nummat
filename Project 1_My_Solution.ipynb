### Problem 1
In the lectures, you learned about normal matrices, those which have a diagonalization of the form

$$
     A = U\Lambda U^H
$$

where $\Lambda$ is a diagonal complex $n\times n$-matrix and $U$ is a unitary (complex) matrix such that $U^H U=I$
(recall that $U^H$ is the complex conjugate of $U^T$).

Show that for any such matrix, one has $\|A\|_2=\rho(A)$ (where $\rho(A)$ is the spectral radius of $A$).

### Solution
Every vector norm induces a norm on matrices $V \in \mathbb{R}^n$, $A \in \mathbb{R}^{n x n}$

$$
\begin{equation}\label{induced_norm}
    ||A|| = \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||}{||\mathbf{x}||}.
\end{equation}
$$

It is also known that 

$$
\begin{equation}\label{vector2norm}
    ||\mathbf{x}||_2 = \left(\sum_{i = 1}^{n} |x_i|\right)^{1/2}.
\end{equation}
$$

A hermitian matrix $B^HB$ can be written as 

$$
\begin{equation}
    U^HB^HBU = \text{diag}(\mu_1, \mu_2, ..., \mu_n), 
\end{equation}
$$

where $\mu_i$ are the positive eigenvalues of $B^HB$ and $U$ is a unitary matrix.  Here, $A = U\Lambda U^H$ is hermitian, since $A^H = A$, and $B^HB$ can be exchanged for the hermitian matrix $A$ in the expression above. This means that $\mu_i$ are the positive eigenvalues of $A$, which means that $\Lambda$ is given by 

$$
\begin{equation}
    \Lambda = \text{diag}(\lambda_1, ..., \lambda_n), 
\end{equation}
$$

where $\lambda_i$ are the eigenvalues of $A$.

Beginning with the first equation above for the spectral norm

$$
\begin{align*}
    ||A||_2 &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \frac{||A\mathbf{x}||_2}{||\mathbf{x}||_2} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A\mathbf{x},A\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(A^HA\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(U\Lambda^2 U^H\mathbf{x},\mathbf{x})}{(\mathbf{x}, \mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(UU^H\mathbf{x}, \mathbf{x})}} &&\text{(Since $U$ is unitary)} \\
    &= \max\limits_{0 \neq \mathbf{x} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 U^H\mathbf{x},U^H\mathbf{x})}{(U^H\mathbf{x}, U^H\mathbf{x})}} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{(\Lambda^2 \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} &&\text{(Substitute $\mathbf{y} = U^H\mathbf{x}$)} \\
    &= \max\limits_{0 \neq \mathbf{y} \in \mathbb{R}^n} \sqrt{\frac{\Lambda^2( \mathbf{y},\mathbf{y})}{(\mathbf{y}, \mathbf{y})}} \\
    &= \max\limits_{i = 1, 2,\ldots, n} \sqrt{\lambda_i^2} \\
    &= \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| \\
    &= \rho (A), 
\end{align*}
$$

which was to be proven. $\, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \,  \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \, \blacksquare$

### Solution
 
**a)** By calculating $\text{det}(A - \lambda I_n) = 0$ one finds that the eigenvalues of $A$ are given by $(-\lambda)^n = 0 \implies \lambda = 0$. This eigenvalue has multiplicity $n$. The Gershgorin circle theorem tells us that all the eigenvalues of $A$ are within at least one of the Gershgorin discs of $A$. A Gershgorin disk $D_i$ is a closed disk with centre given by the diagonal entry in row i, $a_{ii}$, and radius $r_i$ given by the sum of all the entries in row $i$ which are not on the diagonal: $ r_i = \sum_{i \neq j}a_{ij}$. In this specific case this means that all the eigenvalues of $A$ would be contained in a disk with centre 0 and radius 1 and/or in a disk with centre 0 and radius 0. Since at least one of the eigenvalues of the matrix is contained in the disk with centre 0 and radius 0, which corresponds to the point $(0,0)$, at least one eigenvalue must be zero and $A$ must be singular. This way we could have reasoned that $A$ must be singular based on Gershgorin's theorem instead of calculating the eigenvalues of the marix, as we did to begin with.

**b)** Since the task description only says that $\epsilon$ is a small number I will assume that it can be both a positive or a negative, real number. After the new addition to the matrix $A$, $\hat A$ takes the form

$$
\hat A = \left[
\begin{array}{rrrr}
0 & \ldots & \ldots &  \epsilon \\
1 & 0 &   & \vdots \\
\vdots & \ddots &\ddots & \vdots \\
0 & \ldots& 1 & 0 & 
\end{array}
\right].
$$

This leads to different eigenvalues than for the matrix $A$. Computing the eigenvalues as normally gives

$$
    \text{det}(\hat A - \lambda I_n) = 0 \implies (-\lambda)^n + (-1)^{n-1}\epsilon, 
$$

where $I_n$ is the identity matrix in $\mathbb{R}^n$. This ultimately gives 

$$
    \lambda^n = \epsilon. 
$$

This means that the eigenvalues of $\hat A$ are the $n^{th}$ complex roots of epsilon. More specifically, the $m^{th}$ eigenvalue is given by the expression

$$
    \lambda_m = |\epsilon|^{1/n}\exp{\left(\frac{2\pi i m}{n}\right)}\exp{\left(\frac{\theta}{n}\right)}, 
$$

for $1 \leq m \leq n$, where $\theta$ can be $0$ or $\pi$ depending on whether $\epsilon$ is positive or negative. 

Moreover, by the definition of the spectral radius 

$$
    \rho(\hat A) = \max\limits_{i = 1, 2,\ldots, n} |\lambda_i| = |\epsilon|^{1/n},
$$

where I have added the absolute value of $\epsilon$ so that the expression also is correct in a case where $\epsilon < 0$.

Now that the expression for the eigenvalues has been found, an expression for the eigenvectors is given by

$$
    (\hat A -I_n \lambda_i)x_i = 0, 
$$

for every $i = 1, \ldots, n$. This results in the following system

$$
\left[
\begin{array}{rrrr}
-\lambda_i & \ldots & \ldots &  \epsilon \\
1 & -\lambda_i &   & \vdots \\
\vdots & \ddots &\ddots & \vdots \\
0 & \ldots& 1 & -\lambda_i & 
\end{array}
\right] x_i = 0, 
$$

where $x_i = [x_{i_1}, x_{i_2}, \ldots,x_{i_n}]^T \neq 0$ is eigenvector number $i$.

This system gives the following expressions for the coordinates in $x_i$

$$
\begin{align*}
x_{i_1} &= \frac{\epsilon}{\lambda_i}x_{i_n} \\
x_{i_2} = \frac{x_{i_1}}{\lambda_i} &= \frac{\epsilon}{\lambda_i^2}x_{i_n} \\
\vdots \\
x_{i_{n-1}} = \frac{x_{i_{n-2}}}{\lambda_i} &= \frac{\epsilon}{\lambda_i^{n-1}}x_{i_n} \\
x_{i_n} = \frac{x_{i_{n-1}}}{\lambda_i} &= \frac{\epsilon}{\lambda_i^n}x_n = x_{i_n}. \\
\end{align*}
$$

It is apparent that the last coordinate $x_{i_n}$ can be chosen freely. Set $x_{i_n} = \psi \in \mathbb{C}$.
Therefore, the eigenvalue $\lambda_i$ has the corresponding eigenvector $x_i$, given by  

$$
x_i = \left[
\begin{array}{rrrrr}
x_{i_1} \\ x_{i_2} \\ x_{i_3}\\ \vdots \\ x_{i_n}
\end{array}
\right] = 
\left[
\begin{array}{rrrr}
\lambda_i^{-1} \\ \lambda_i^{-2} \\ \lambda_i^{-3} \\ \vdots \\ \epsilon^{-1}
\end{array}
\right]\epsilon \psi,
$$

for every $i = 1, \ldots, n$. This is the expression for the eigenvectors of $\hat A$. More specifically, the $m^{th}$ component of the $j^{th}$ eigenvector is given by

$$
    x_{j_m} = \left(\epsilon^{1/n}\exp{\left(\frac{2\pi i j}{n}\right)}\exp{\left(\frac{\theta}{n}\right)}\right)^{-m}\epsilon \psi = \epsilon^{-m/n+1}\exp{\left(-\frac{2\pi i j m}{n}\right)}\exp{\left(-\frac{m\theta}{n}\right)}\psi, 
$$

where the same restrictions on $\theta$ as before apply. 

**c)**
It is known that $K_2(\hat{A})=\|\hat{A}\|_2\cdot\|\hat{A}^{-1}\|_2 = \frac{\sigma_1(\hat A)}{\sigma_n(\hat A)}$, where $\sigma_1{(\hat A)}$ and $\sigma_n{(\hat A)}$ are the maximum and minimum singular values of $\hat A$. The singular values are defined as $\sigma_i{(\hat A)} = \sqrt{\lambda_i (\hat A\hat A^H)}$, for $i = 1, \ldots, n$. The product $\hat A \hat A^H$ is given by

$$
\hat A \hat A^H= \left[
\begin{array}{rrrr}
0 & \ldots & \ldots &  \epsilon \\
1 & 0 &   & \vdots \\
\vdots & \ddots &\ddots & 0 \\
0 & \ldots& 1 & 0 & 
\end{array}
\right]\left[
\begin{array}{rrrr}
0 & 1 & \ldots &  0 \\
\vdots & 0 & \ddots & \vdots \\
\vdots &  & \ddots & 1 \\
\epsilon & \ldots & 0 & 0 & 
\end{array}
\right] = \left[
\begin{array}{rrrr}
\epsilon^2 & 0 & \ldots &  0 \\
0 & 1 & \ddots & \vdots \\
\vdots &  & \ddots & 0 \\
0 & \ldots & 0 & 1 & 
\end{array}
\right].
$$

Since $\hat A \hat A^H$ is a diagonal matrix, the eigenvalues are given by the diagonal entries of the matrix. Therefore, $\sqrt{\lambda_i (\hat A\hat A^H)} = |\epsilon|$ or $\sqrt{\lambda_i (\hat A\hat A^H)} = 1$ for all $i$. Which of these values is the maximum and minimum, depends on the size of $\epsilon$ compared to $1$. This gives the following expression for the condition number $K_2(\hat A)$

$$
    K_2(\hat A) = \begin{cases} 
      |\epsilon| & |\epsilon| > 1 \\
      1 & \epsilon = 1 \\
      \frac{1}{|\epsilon|} & |\epsilon| < 1 
       \end{cases}.
$$

**b)**

Define the expression 

$$
    \Xi_i = \frac{\frac{\|\delta x\|_i}{\|x\|_i}}{\frac{\|\delta H_n\|_i}{\|H_n\|_i}+\frac{\|\delta b\|_i}{\|b\|_i}},
$$

for $i = 1, 2, \infty$.



```python
import numpy.random as r
import numpy.linalg as l

# Using the same n-values as in a)
n = np.arange(2, 14, 1, dtype=int)

hmin = lambda n: 1/(2*n+1) # Smallest value in H_n.
pert_amount = 2000 # Amount of perturbations.

# Empty dictionary to store values per value of n.
simulations = {}

def run_sim():
    """Run simulation and fill dictionary 'simulations' with data."""
    for i in range(n[-2]):
        s = n[i]
        b = np.ones(s)
        H = la.hilbert(s)
        x = l.solve(H,b)
        values = np.zeros((pert_amount, 4))

        c = 1e-15 # Factor that scales the perturbations. Should be a small perturbation.
        # Went to the smallest possible order of magnitude before encountering overflow.

        j = hmin(s)*c # Constant for scaling perturbations in H_n.

        for p in range(pert_amount):
            dH = r.rand(s,s)*j 
            db = r.rand(s)*c
            x_dx = l.solve(H+dH, b+db)
            dx = x - x_dx # Find perturbation in x.

            frob_norm = (l.norm(dx)/l.norm(x))/(((l.norm(dH))/l.norm(H))+((l.norm(db)/l.norm(b))))
            inf_norm = (l.norm(dx, np.inf)/l.norm(x, np.inf))/(((l.norm(dH, np.inf))/l.norm(H, np.inf))+((l.norm(db, np.inf)/l.norm(b, np.inf))))
            one_norm = (l.norm(dx, 1)/l.norm(x, 1))/(((l.norm(dH, 1))/l.norm(H, 1))+((l.norm(db, 1)/l.norm(b, 1))))

            # For plotting values for 3 common norms and ||dx||_2.
            val = [l.norm(dx), frob_norm, inf_norm, one_norm]
            values[p, :] = val

        simulations[i] = values # One list of values for the perturbations, per n.
    

# Used to plot some averages below.
average_diffs = np.zeros(len(n))
average_frob_norm = np.zeros(len(n))
average_inf_norm = np.zeros(len(n))
average_one_norm = np.zeros(len(n))
    
def find_average_values(log = False):
    """Fill arrays with average of fractions and diffs per n. 
    
    Added capabilities to print the simulated values (console log) if 'log' == True. 
    """
    for key, arr in simulations.items():
        # Sum for plotting averages for each n.
        sum_diffs = np.sum(arr, axis = 0)[0]
        sum_frob_norm = np.sum(arr, axis = 0)[1]
        sum_inf_norm = np.sum(arr, axis = 0)[2]
        sum_one_norm = np.sum(arr, axis = 0)[3]
        
        average_diffs[key] = sum_diffs/pert_amount
        average_frob_norm[key] = sum_frob_norm/pert_amount
        average_inf_norm[key] = sum_inf_norm/pert_amount
        average_one_norm[key] = sum_one_norm/pert_amount
        
        if log:
            print("Sim for n =", n[key], ":")
            for a in arr:
                print("||dx|| = ", a[0], ",fraction = ", a[1])
                print("\n")
            print("\n\n")

run_sim()
find_average_values()

fig, ax = plt.subplots(1,3, figsize=plt.figaspect(1/4))
fig.suptitle("Some Averages Over "+str(pert_amount)+" Perturbations per Value of $n$")
ax[0].plot(n, average_diffs)
ax[0].set_title("Average of $||\delta x||_2$ Over Perturbations as Function of $n$")
ax[0].set_xlabel("Dim [n]")
ax[0].set_ylabel(r"$\overline{||\delta x||}_2$")
ax[1].plot(n, average_one_norm, label="$\overline{\Xi}_1$")
ax[1].plot(n, average_frob_norm, label="$\overline{\Xi}_2$")
ax[1].plot(n, average_inf_norm, label="$\overline{\Xi}_{\infty}$")
ax[1].set_title("Average of $\Xi_i, \, \, \, i = 1, 2, \infty$, as Function of $n$")
ax[1].set_xlabel("Dim [n]")
ax[1].set_ylabel("$\overline{\Xi}_i, \, \, \, i = 1, 2, \infty$")
ax[1].legend(loc=2)
ax[2].plot(n, np.log(average_one_norm), label="$\log{(\overline{\Xi}_1)}$")
ax[2].plot(n, np.log(average_frob_norm), label="$\log{(\overline{\Xi}_2)}$")
ax[2].plot(n, np.log(average_inf_norm), label="$\log{(\overline{\Xi}_{\infty})}$")
ax[2].set_title("$\log{(\overline{\Xi}_i)}, \, \, \, i = 1, 2, \infty$, as Function of $n$")
ax[2].set_xlabel("Dim [n]")
ax[2].set_ylabel("$\log{(\overline{\Xi}_i)}, \, \, \, i = 1, 2, \infty$")
ax[2].legend(loc=2)
plt.show()
```

### Problem X

Let $A$ be any invertible $n\times n$-matrix. Suppose that $\delta A$ is the smallest possible matrix, measured in a subordinate (natural) matrix norm $\|\cdot\|$ such that $A+\delta A$ is singular. Show that

$$
    \|\delta A\|  = \|A^{-1}\|^{-1}
$$



### Solution

This proof is divided into two parts. In the first part it will be proved that a matrix $A+\delta A$ only can be singular if $\|A^{-1}\|^{-1} \leq \|\delta A\|$. In the second part, the equality will be proved, by showing that a matrix $\delta A$ can be constructed such that $A+\delta A$ is singular and its norm is given by $\|\delta A\|  = \|A^{-1}\|^{-1}$.

Firstly, since $A$ is invertible and $A+\delta A$ is singular, the following must hold for some $x \neq 0$

$$
    (A+\delta A)x = 0 \iff Ax + \delta Ax = 0 \iff -\delta Ax = Ax \iff x = -A^{-1}\delta A, 
$$

where the last equality holds since $A$ is invertible. 

Taking the norm of this result one gets

$$
    \|x\| = \|-A^{-1}\delta Ax\| \leq \|A^{-1}\|\delta Ax\| \leq \|A^{-1}\|\delta A\|\|x\|, 
$$

where we assume that the matrix norm has the sub-multiplicative property $\|AB\| \leq \|A\|\|B\|$ and that the matrix norm and the vector norm are compatible $\|Ax\| \leq \|A\|\|x\|$. The relation therefore is

$$
    \|A^{-1}\|^{-1} \leq \|\delta A\|.
$$

It has therefore been shown that the following equivalence holds 

$$
    A + \delta A \text{ singular} \iff \|A^{-1}\|^{-1} \leq \|\delta A\|.
$$

Secondly, a candidate for $\delta A$ shall be constructed so that the two necessary properties presented in the beginning hold. The candidate that will be used is

$$
    \delta A = -\|A^{-1}\|^{-1}xy^T, 
$$

where 

$$
\begin{align*}
\|x\| = 1, \, && \|y\|_* :=\max_{z \neq 0} \frac{|y^Tz|}{\|z\|} = 1, && \|A^{-1}\| = y^TA^{-1}x. 
\end{align*}
$$

It will be proven that, given the constructed matrix $\delta A$, $A + \delta A$ is singular, which is one of the properties that needs to be fulfilled. This will be done by computing

$$
    (A+\delta A)\cdot A^{-1}x.
$$

For $A+\delta A$ to be singular the equation above must have a solution for some $x \neq 0$, since this would show that $(A+\delta A)\cdot A^{-1}$ is singular. Since $A^{-1}$ is not singular, the other factor must be singular. The equation gives 

$$
    (A+\delta A)\cdot A^{-1}x = (I-\|A^{-1}\|^{-1}xy^TA^{-1})x = Ix-\|A^{-1}\|^{-1}xy^TA^{-1}x = Ix-\|A^{-1}\|^{-1}x\|A^{-1}\| = (I-I)x = 0, 
$$

where we have used the defined norm of $A^{-1}$. This means that the equation $(A+\delta A)\cdot A^{-1}x = 0$ has a solution for all $x$, which implies that $A+\delta A$ must be singular. Therefore, our construction of matrix $\delta A$ fulfills the property that $A + \delta A$ should be singular.

Next, we shall prove that the norm of this constructed matrix is given by $\|\delta A\|  = \|A^{-1}\|^{-1}$. The proof reads

$$
    \|\delta A\| = \max\limits_{0 \neq \mathbf{z} \in \mathbb{R}^n} \frac{\|\delta Az\|}{\|z\|} = \max\limits_{0 \neq \mathbf{z} \in \mathbb{R}^n} \frac{\|A^{-1}\|^{-1}\|xy^Tz\|}{\|z\|} \leq \|A^{-1}\|^{-1} \|x\| \max\limits_{0 \neq \mathbf{z} \in \mathbb{R}^n} \frac{|y^Tz|}{\|z\|} = \|A^{-1}\|^{-1},
$$

where we have assumed sub-multiplicativity of the matrix norms as well as compatible vector and matrix norms. We have also used the defined norms of $x$ and $y$ that appear in $\delta A$. Since we know that $A + \delta A$ is singular, we know, from part one of the proof, that 

$$
    \|A^{-1}\|^{-1} \leq \|\delta A\|.
$$

From this it follows that 

$$
    \|\delta A\| \leq \|A^{-1}\|^{-1} \leq \|\delta A\| \implies \|A^{-1}\|^{-1} = \|\delta A\|.
$$

Thus, the problem has been proved given the constructed matrix $\delta A$. The only thing remaining to prove is existence of such vectors $x$ and $y$ according to the construction of  $\delta A$.

From the definition of a subordinate norm we know that

$$
    \|A^{-1}\| = \max_{\|x'\| = 1} \|A^{-1} x'\|, 
$$

for a vector $x'$.

This implies that 

$$
    \exists \text{ vector }x : \|A^{-1} x\| = \|A^{-1}\|. \text{ with } \|x\| = 1.
$$

With this $x$ we can define 

$$
    \nu = \frac{A^{-1}x}{\|A^{-1}x\|}, 
$$

which implies that 

$$
    \|\nu\| = \frac{\|A^{-1}x\|}{\|A^{-1}x\|} = 1. 
$$

Now, one can define an induced norm given by

$$
    \|\xi\|_{**} :=\max_{y' \neq 0} \frac{|\xi^Ty'|}{\|y'\|_*} = \max_{\|y'\|_* = 1} |\xi^Ty'|,
$$

which is induced by a norm $\|\cdot\|_*$. In a finite dimensional vector space we know that 

$$
    \|y\|_{**} = \|y\|
$$

for any vector $y$. This means that $\|y\|_{**} = 1$, since we stated that $\|\nu\| = 1$ above, and the maximum when $\|y'\|_* = 1$ exists, which ultimately means that $\|y'\|_* = 1$ for any vector $y'$.   

Therefore, 

$$
    \|y\|_* :=\max_{\nu \neq 0} \frac{|y^T\nu|}{\|\nu\|} = \max_{\|\nu\| = 1} |y^T\nu| = 1. 
$$

This means that the norms $\|x\| = 1$ and $\|y\|_* = 1$ stated together with the constructed matrix $\delta A$ exist. Moreover. it needs to be proved that $\|A^{-1}\| = y^TA^{-1}x$. Since A is invertible, we can define a vector $\xi$ such that $\xi = A^{-1}x$. From what we just proved it follows that

$$
    1 = \|y\|_* = \max_{\xi \neq 0} \frac{|y^T\xi|}{\|\xi\|} = \max_{x \neq 0} \frac{|y^TA^{-1}x|}{\|A^{-1}x\|} = \max_{x \neq 0}\frac{|y^TA^{-1}x|}{\|A^{-1}\|\|x\|},
$$

from the assumption that the matrix and vector norms are compatible, as well as the following statement, that was proved earlier 

$$
    \exists \text{ vector }x : \|A^{-1} x\| = \|A^{-1}\| = \|A^{-1}\|\|x\| \text{ with } \|x\| = 1.
$$

This means that 

$$
    \|A^{-1}\| = \max_{\|x\| = 1} |y^TA^{-1}x|. 
$$

It has therefore been shown that 

$$
    \exists \text{ vector }x : \|A^{-1}\| = y^TA^{-1}x \text{ with } \|x\| = 1, 
$$

and the proof is complete.
